## **Mini-Project: Single Product E-commerce Calculator**

Welcome to your basic inventory tool! In this project, you'll build a Python program to calculate the cost of a *single product*, including GST. This will help you master **variables**, **datatypes**, **user input**, and **arithmetic operators**.

### **Project Requirements:**
*   **Welcome Message**: A simple greeting for the calculator.
*   **User Inputs**: Ask the user to input details for ONE product:
    *   `product_name` (String)
    *   `product_count` (Integer)
    *   `price_per_item` (Float)
    *   `gst_percentage` (Float - for example, 18 for 18%)
*   **Data Type Verification**: Use `type()` to print and confirm the data types of all inputs.
*   **Calculations**: Use arithmetic operators to find:
    *   `amount_before_gst` (count * price_per_item)
    *   `gst_amount` (amount_before_gst * (gst_percentage / 100))
    *   `amount_after_gst` (amount_before_gst + gst_amount)
*   **Summary Report**: Print a clean summary of the product's details and all calculated costs.

### **Task 1: Welcome Message and Product Inputs**

Start by greeting the user and then ask for all the details about one product.

**Your Turn:**
1.  Print a welcome message for the calculator.
2.  Create a variable `product_name` and assign it a string value from user input.
3.  Create a variable `product_count` and assign it an integer value from user input.
4.  Create a variable `price_per_item` and assign it a float value from user input.
5.  Create a variable `gst_percentage` and assign it a float value from user input.

```python
# Write your code for the welcome message and user inputs here

print("Welcome to the Single Product E-commerce Calculator!")
print("Let's calculate the cost for one item.")

product_name = input("Enter the product name: ")
product_count = int(input("Enter the quantity of the product: "))
price_per_item = float(input("Enter the price per item: "))
while True:
     gst_input = input("Enter the GST percentage: ")
     try:
          gst_percentage = float(''.join(ch for ch in gst_input if ch.isdigit() or ch in '.-'))
          break
     except ValueError:
          print("Please enter a valid numeric GST percentage (for example, 18).")
```

### **Task 2: Data Type Verification**

It's a good practice to verify that your variables have the correct data types after conversion. Use the `type()` function for this.

**Your Turn:**
1.  Print the data type of `product_name`.
2.  Print the data type of `product_count`.
3.  Print the data type of `price_per_item`.
4.  Print the data type of `gst_percentage`.

```python
# Write your code for data type verification 
print("\n--- Data Type Verification ---")
print("Data type of product_name:", type(product_name))
print("Data type of product_count:", type(product_count))
print("Data type of price_per_item:", type(price_per_item))
print("Data type of gst_percentage:", type(gst_percentage))
```

### **Task 3: Calculations**

Now, perform the necessary calculations using arithmetic operators to find the costs before and after GST.

**Your Turn:**
1.  Calculate `amount_before_gst` (product_count * price_per_item).
2.  Calculate `gst_amount` (amount_before_gst * (gst_percentage / 100)).
3.  Calculate `amount_after_gst` (amount_before_gst + gst_amount).

```python
# Write your code for calculations here

# Task 3: Calculations
amount_before_gst = product_count * price_per_item
gst_amount = amount_before_gst * (gst_percentage / 100)
amount_after_gst = amount_before_gst + gst_amount

# Output
print("\n--- Calculation Results ---")
print(f"Amount before GST: {amount_before_gst}")
print(f"GST Amount ({gst_percentage}%): {gst_amount}")
print(f"Amount after GST: {amount_after_gst}")
```

### **Task 4: Display Summary Report**

Finally, present all the product details and calculated costs in a clear summary report. Use f-strings for neat formatting, especially for currency values.

**Your Turn:**
1.  Print the `product_name`.
2.  Print the `product_count`.
3.  Print the `price_per_item`.
4.  Print the `gst_percentage` applied.
5.  Print the `amount_before_gst`.
6.  Print the `gst_amount`.
7.  Print the `amount_after_gst`.

```python
# Write your code for the summary report here

print("\n--- Summary Report ---")
print(f"Product Name       : {product_name}")
print(f"Quantity           : {product_count}")
print(f"Price per Item     : PKR {price_per_item:.2f}")
print(f"GST Percentage     : {gst_percentage:.2f}%")
print(f"Amount Before GST  : PKR {amount_before_gst:.2f}")
print(f"GST Amount         : PKR {gst_amount:.2f}")
print(f"Amount After GST   : PKR {amount_after_gst:.2f}")
```
